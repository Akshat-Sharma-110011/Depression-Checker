```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn.impute import SimpleImputer
import pandas as pd

# Assuming final_df is your DataFrame
X = df.drop(columns=['Depression'])  # Features
y = df['Depression']  # Target

# Define categorical columns that need to be encoded
categorical_cols = ['Sleep Duration', 'Gender', 'City', 'Profession', 'Dietary Habits', 'Degree', 'Suicidal Thoughts', 'Depression History']  # Add all categorical columns

# Create the preprocessor to handle categorical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), X.select_dtypes(include=['int64', 'float64']).columns),  # Scaling numerical columns
        ('cat', OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore'), categorical_cols)  # OneHotEncoding for categorical columns
    ])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)

# Define the pipeline with preprocessor and AdaBoostClassifier
pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', AdaBoostClassifier(n_estimators=500, learning_rate=0.07, random_state=3))
])

# Fit the pipeline on training data
pipe.fit(X_train, y_train)
```
